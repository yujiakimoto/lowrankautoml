```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.preprocessing import LabelEncoder
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_csv('error_matrix_small.csv', index_col=0)
models = np.array(list(df))
dataset_IDs = df.index.values
errorMtx = df.values
n_datasets, n_models = errorMtx.shape
```

```python
def sameAlg(testBestCol):
    testBestAlgType = []
    for i in range(len(testBestCol)):
        # testBestAlgType.append(eval(models[testBestCol[i]])['algorithm'])
        testBestAlgType.append(models[testBestCol[i]].split('.')[0])
    return np.array(testBestAlgType)
```

```python
bestModel = np.argmin(errorMtx, axis=1)
bestAlg = sameAlg(bestModel)
bestError = np.min(errorMtx, axis=1)
algtypes = np.sort(np.unique(bestAlg))
```

```python
labels = LabelEncoder().fit_transform(bestAlg)
colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 
          'tab:brown', 'tab:pink', 'tab:gray', 'tab:olive', 'tab:cyan']
```

```python
# 2-dimensional t-SNE embedding
embedded = TSNE().fit_transform(errorMtx)
```

```python
plt.figure(figsize=(10,5))

plots = []
for i in range(10):
    idx = np.where(labels == i)[0]
    plots.append(plt.scatter(embedded[idx,0], embedded[idx,1], color=colors[i]))

plt.legend(tuple(plots), tuple(algtypes), loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
# 3-dimensional t-SNE embedding
embedded = TSNE(n_components=3).fit_transform(errorMtx)
```

```python
fig = plt.figure(figsize=(10,5))
ax = plt.subplot(111, projection='3d')

for i in range(10):
    idx = np.where(labels == i)[0]
    ax.plot(embedded[idx,0], embedded[idx,1], embedded[idx,2], 'o', color=colors[i], label=algtypes[i])

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
